```python
import pandas as pd
import numpy as np
from sklearn import preprocessing as pp
from tensorflow.keras.utils import to_categorical
from tensorflow.keras import layers
from tensorflow.keras import models
```

<h1 style='text-align:center;'>Load Data Set</h1> 

```python
dataFrame = pd.read_csv('sonar.csv')
dataFrame.head()
```

<h1 style='text-align:center;'>Shuffle Data</h1>

```python
dataFrame_shuffle = dataFrame.sample(frac=1)
```

<h1 style='text-align:center;'>Encoding Labels</h1>

```python
labels = dataFrame_shuffle.iloc[:,60]
encoder = pp.LabelEncoder()
encoder.fit(labels)
encoded_label = encoder.transform(labels)
encoded_label
```

<h1 style='text-align:center;'>Spliting Inputs and Labels</h1>

```python
train_data = dataFrame_shuffle.iloc[:144,:60].values
df_train_label = encoded_label[:144]
test_data = dataFrame_shuffle.iloc[144:,:60].values
df_test_label = encoded_label[144:]
```
